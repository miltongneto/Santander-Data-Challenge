<h2>Dados referentes a Pesquisa Mensal de Comércio </h2>

Índice de volume de vendas no comércio varejista, olhando a produção do mês referente ao ano passado, já com um ajuste de sazonalidade.
&nbsp;

*Fonte: Sidra - IBGE (https://sidra.ibge.gov.br/tabela/3418) - Maio2019/Maio2020 - última atualização: Jul2020*

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pesquisa_mensal_comercio.csv',sep=';',skiprows = 3)
```

```python
df.head()
```

```python
df.drop(index=0)
```

```python
df.info()
```

```python
# Selecionando no contexto do Brasil
df_brasil = df[df['Nível']=='BR']
df_brasil.head()
```

```python
# Selecionando no contexto de cada UF
df_uf = df[df['Nível']=='UF']
df_uf.head()
```

```python
print(df_brasil.columns)
```

```python
# Pivotando as informações dos meses - Brasil

df_brasil_pivot = df_brasil.melt(id_vars = ['Atividades','Nível','Brasil e Unidade da Federação'], 
                                 var_name = 'Mes Variacao', 
                                 value_name = 'Variacao')
```

```python
# transformando para float
df_brasil_pivot['Variacao'] = df_brasil_pivot['Variacao'].str.replace(',','.').astype(float)
```

```python
df_brasil_pivot
```

```python
# Pivotando as informações dos meses - UF

df_uf_pivot = df_uf.melt(id_vars = ['Atividades','Nível','Brasil e Unidade da Federação'], 
                                 var_name = 'Mes Variacao', 
                                 value_name = 'Variacao')
```

```python
# transformando para float
df_uf_pivot['Variacao'] = df_uf_pivot['Variacao'].str.replace(',','.').astype(float)
```

```python
df_uf_pivot
```

```python
df_brasil_pivot.head()
```

```python
# Buiscando os tipos distintos de atividades 

df.Atividades.unique()
```

No total, vemos que temos **9** tipos de macro atividades mapeadas pelo Sidra:

1. Combustíveis e lubrificantes
- Hipermercados e supermercados
- Tecidos, vestuário e calçados
- Móveis 
- Eletrodomésticos
- Artigos farmacêuticos, médicos, ortopédicos, de perfumaria e cosméticos
- Livros, jornais, revistas e papelaria
- Equipamentos e materiais para escritório, informática e comunicação
- Outros artigos de uso pessoal e doméstico

<br>

Com base nisso, vamos plotar a nivel *UF* e *Brasil* o histórico de cada atividade e ver qual teve uma maior queda.

```python
df_brasil_combustiveis = df_brasil_pivot[df_brasil_pivot['Atividades']=='Combustíveis e lubrificantes']
df_brasil_mercado = df_brasil_pivot[df_brasil_pivot['Atividades']=='Hipermercados e supermercados']
df_brasil_vestuario = df_brasil_pivot[df_brasil_pivot['Atividades']=='Tecidos, vestuário e calçados']
df_brasil_moveis = df_brasil_pivot[df_brasil_pivot['Atividades']=='Móveis']
df_brasil_eletrodomestico = df_brasil_pivot[df_brasil_pivot['Atividades']=='Eletrodomésticos']
df_brasil_farmaceutico = df_brasil_pivot[df_brasil_pivot['Atividades']=='Artigos farmacêuticos, médicos, ortopédicos, de perfumaria e cosméticos']
df_brasil_livros = df_brasil_pivot[df_brasil_pivot['Atividades']=='Livros, jornais, revistas e papelaria']
df_brasil_escritorio = df_brasil_pivot[df_brasil_pivot['Atividades']=='Equipamentos e materiais para escritório, informática e comunicação']
df_brasil_outros = df_brasil_pivot[df_brasil_pivot['Atividades']=='Outros artigos de uso pessoal e doméstico']
```

<h4> Conclusões </h4>

Analisando a evolução das variações entre os meses e as atividades, é possível colocar alguns pontos importantes:

1. Nem todos os setores de atividades passaram por problemas devido a pandemia. Um exemplo desse caso é o setor de compras feitas em supermercados (que tiveram em 2020 apenas o mês de Janeiro com queda comparado com Jan2019) e o de produtos farmacêuticos (que tiveram queda em Abril e Maio, porém a partir de Maio é possível ver uma melhora).
2. Podemos perceber que diferente dos setores apresentados acima, alguns começaram a sofrer e ainda estão em queda desde Fev2020, como por exemplo é o caso do segmento de Livros e Jornais.
3. Olhando para um segmento em específico, o de equipamentos e materiais de escritório, podemos ver uma queda de Jan2020 a Abr2020 e isso se relaciona com o fato de que em com o passar do tempo, as empresas gastaram menos com relação ao compra desse tipo de material devido ao aumento da quantidade dos funcionários em home office.
