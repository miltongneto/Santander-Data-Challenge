<h2> Pesquisa Mensal de Comércio </h2>


Visto que estamos no cenário de pandemia, é importante entender como caminha a economia do Brasil, analisando o volume de vendas de cada mês, dentro de cada macro atividade.

Nessa análise, vamos explorar como o comércio varejista está em cada macro atividade, comparando os pontos percentuais de variação do mês analisado contra o mesmo mês no ano anterior.

Fonte: Sidra - IBGE (https://sidra.ibge.gov.br/tabela/3418) - Maio2018/Maio2020 - última atualização: Jul2020


```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Pesquisa Mensal de Comércio.csv',sep=';',skiprows = 3)
```

```python
df
```

```python
df.drop(index=0)
```

```python
df.info()
```

```python
print(df.columns)
```

```python
df1 = df[df['Nível'] == 'BR']
df1.head()
```

```python
# Pivotando as informações dos meses - UF

df_pivot = df1.melt(id_vars = ['Atividades','Nível','Unnamed: 2'], 
                                 var_name = 'Mes Variacao', 
                                 value_name = 'Variacao')
```

```python
df_pivot
```

```python
df2 = df_pivot.rename(columns={'Unnamed: 2': 'Brasil'})
```

```python
df2
```

```python
# transformando para float
df2['Variacao'] = df2['Variacao'].str.replace(',','.').astype(float)
```

```python
df2.head()
```

```python
# Buiscando os tipos distintos de atividades 

df2.Atividades.unique()
```

No total, vemos que temos **9** tipos de macro atividades mapeadas pelo Sidra:

1. Combustíveis e lubrificantes
- Hipermercados e supermercados
- Tecidos, vestuário e calçados
- Móveis 
- Eletrodomésticos
- Artigos farmacêuticos, médicos, ortopédicos, de perfumaria e cosméticos
- Livros, jornais, revistas e papelaria
- Equipamentos e materiais para escritório, informática e comunicação
- Outros artigos de uso pessoal e doméstico

<br>

Com base nisso, vamos plotar o histórico de cada atividade e ver qual teve uma maior queda.

```python
df2_combustiveis = df2[df2['Atividades']=='Combustíveis e lubrificantes']
df2_mercado = df2[df2['Atividades']=='Hipermercados e supermercados']
df2_vestuario = df2[df2['Atividades']=='Tecidos, vestuário e calçados']
df2_moveis = df2[df2['Atividades']=='Móveis']
df2_eletrodomestico = df2[df2['Atividades']=='Eletrodomésticos']
df2_farmaceutico = df2[df2['Atividades']=='Artigos farmacêuticos, médicos, ortopédicos, de perfumaria e cosméticos']
df2_livros = df2[df2['Atividades']=='Livros, jornais, revistas e papelaria']
df2_escritorio = df2[df2['Atividades']=='Equipamentos e materiais para escritório, informática e comunicação']
df2_outros = df2[df2['Atividades']=='Outros artigos de uso pessoal e doméstico']
```

```python
df2_combustiveis.dtypes
```

```python
df2_combustiveis.plot(x='Mes Variacao', y='Variacao', figsize=(15,5),
                            title='Variação (p.p) do volume de vendas - Combustíveis e Lubrificantes',
                            marker='o')
```

<h4> Conclusões </h4>

Analisando a evolução das variações entre os meses e as atividades, é possível colocar alguns pontos importantes:

1. Olhando o histórico e através da evolução da variação, podemos ver que de fato os macro setores de bens de consumo de fato sofreram grandes impactos causados pela crise do novo coronavírus.
2. Nem todos os setores de atividades passaram por problemas devido a pandemia. Um exemplo desse caso é o setor de compras feitas em supermercados (que tiveram em 2020 apenas o mês de Janeiro com queda comparado com Jan2019) e o de produtos farmacêuticos (que tiveram queda em Abril e Maio, porém a partir de Maio é possível ver uma melhora).
3. Podemos perceber que diferente dos setores apresentados acima, alguns começaram a sofrer e ainda estão em queda desde Fev2020, como por exemplo é o caso do segmento de Livros e Jornais.
4. Olhando para um segmento em específico, o de equipamentos e materiais de escritório, podemos ver uma queda de Jan2020 a Abr2020 e isso se relaciona com o fato de que em com o passar do tempo, as empresas gastaram menos com relação ao compra desse tipo de material devido ao aumento da quantidade dos funcionários em home office.
